```python
import pandas as pd
import datetime
```

```python
bites_file = "animal_bites.csv"
bites_data = pd.read_csv(bites_file)

#bites_data_count = len(bites_data['victim_zip'].unique())

#bites_data_count

bites_data.dtypes
```

```python
zip_code_count = bites_data.groupby(['bite_convert_date ']).agg({'SpeciesIDDesc': 'count'}).reset_index()


most_bites = zip_code_count.sort_values('SpeciesIDDesc', ascending= False)


most_bites['year'] = pd.DatetimeIndex(most_bites['bite_convert_date ']).year


most_bites.set_index('year')
most_bites.drop(columns="bite_convert_date ")
```
