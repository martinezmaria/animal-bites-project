```python

read_file= pd.read_csv("Resources/datasets_2515_4202_Health_AnimalBites.csv")
read_file.head()
```

```python
#converting to datatime to be able to extrect year, month, days, and hours 

read_file["bite_convert_date "]= pd.to_datetime(read_file.iloc[:,0], errors="coerce")

read_file
```

```python
lf=read_file.loc[(read_file.iloc[:,15]>= "2007-01-01 00:00:00") & (read_file.iloc[:,15] <= "2017-12-31 00:00:00")]
lf
```

```python
#selecting the needed columns using the loc function
Animal_Bites_Table=lf.iloc[:,[16,1,2,3,4,7]]
Animal_Bites_Table
```

```python
tf= pd.DataFrame(Animal_Bites_Table)
rn= tf.rename(columns={"bite_year":"years","SpeciesIDDesc":"species","BreedIDDesc":"Breed","GenderIDDesc":"Gender","color":"Color","victim_zip":"ZipCode"})
kf= rn.loc[(rn.iloc[:,0]>=2007.0) & (rn.iloc[:,0])<=2017.0]
kf
```

```python
#retrieving  the the number and type of animal from the dataframe 
data_test=kf["species"].value_counts()
data_test
```

```python
#pie chart 
explode=(.1,.1,1,0,.4,.7)
colors=("red","lightcoral","green","lightskyblue","orange","yellow" )
data_test.head(6).plot.pie(colors=colors, autopct="%.1f%%", explode=explode, startangle=140, figsize=(5,5))
plt.show()
```

```python
#accordintg to our Data we can say that dogs are more likely to bite human, they represent 79.2% of all cases follow by cats with 17.7%
#we also negligeable cases like FERRET,RABBIT,SKUNK
```

```python
#find the area location with more bites
#grouping and sorting data 
Animal_Bites_grouped= Animal_Bites_Table.groupby(["victim_zip"])["SpeciesIDDesc"].count().reset_index()
Animal_Bites_sort= Animal_Bites_grouped.sort_values(["SpeciesIDDesc"], ascending=False)
Animal_Bites_sort
```

```python
#Find the Zipcode with the most bite by using the loc function 
most_bite_zip= Animal_Bites_sort.loc[Animal_Bites_sort["SpeciesIDDesc"]==371]
#convert to a dataframe visualisation 
visual_mbz= pd.DataFrame(most_bite_zip)
visual_mbz_rename=visual_mbz.rename(columns={"SpeciesIDDesc":"Number of Bites (max)"})
visual_mbz_rename.reset_index()
```

```python
#
group_Breed_zip= Animal_Bites_Table.groupby(["BreedIDDesc"])["victim_zip"].count().reset_index()
group_Breed_zip_sort= group_Breed_zip.sort_values(["victim_zip"],ascending=False)
bite_per_breed= group_Breed_zip_sort.set_index("BreedIDDesc")
bite_per_breed
```

```python
#barplot
bite_per_breed.head(20).plot(kind= "bar",color="b", figsize=(20,5))
plt.title("Number of bite per dog breed")
plt.xlabel("Dog Breed")
plt.ylabel("number of Bites")
plt.show()
# As we can see, the number of bites is in some way related to the breed of animals 
# therefore if our hypothesis comes to be  true, then we will conclude that one breed is more likely to be prone to violence than 
#another one if exposed at a violent environment 
```
